```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import cv2 as cv    # opneCV
from sklearn import utils
from tqdm.notebook import tqdm    # progressbar처리
import os   # 파일의 경로 설정을 위해 필요
```

```python
# 폴더 경로 설정
train_dir = 'C:/jupyter_home/data/cat_dog/train/train'

# label(target)을 알아내기 위한 함수
# cat = 0 , dog = 1
# img는 파일이름(cat.0.jpg)
def labeling(img):
    class_name = img.split('.')[0]
    if class_name == 'cat':
        return 0
    if class_name == 'dog':
        return 1
```

```python
img = 'cat.0.jpg'
tmp = img.split('.')
print(tmp)
```

```python
# label data와 pixel data(x_data)를 저장할 변수 생성
x_data = []
t_data = []
```

```python
# 파일을 하나씩 반복하면서 처리
# 특정 폴더에 있는 모든 파일명 가져오기 => os.listdir()
for img in tqdm(os.listdir(train_dir),
                total=len(os.listdir(train_dir)),
                position=0,
                leave=True):
    label_data = labeling(img)
    img_path = os.path.join(train_dir,img)
    
    # img_path(이미지 full path)를 이용해서 opencv를 통해 픽셀데이터를 추출
    # cnn을 이용하려면 이미지가 같은 형태여야해서 이미지 resize를 해야함
    img_data = cv.resize(cv.imread(img_path, cv.IMREAD_GRAYSCALE),(80,80))
    # print(img_data.shape)  #(374, 500)
    
    t_data.append(label_data)
    x_data.append(img_data.ravel())
```

```python
t_df = pd.DataFrame({'label':t_data})
display(t_df.head)
```

```python
x_df = pd.DataFrame(x_data)
display(x_df.head())
```

```python
# x_data와 t_data 결합
df = pd.merge(t_df,x_df, left_index=True, right_index=True)
display(df.head())
```

```python
# shuffle을 이용해서 
shuffle_df = utils.shuffle(df)
display(shuffle_df.head())
```

```python
# 최종적으로 만들어진 dataframe을 파일로 저장
shuffle_df.to_csv('C:/jupyter_home/data/cat_dog/train.csv', index=False)
```
