### indexing 

```python
my_dict = {'이름':['홍길동','아이유','김연아','신사임당'],
           '학과': ['철학','수학','컴퓨터','국어국문'],
           '학년':[1,3,2,4],
           '학점':[1.4,2.7,3.5,2.9]}

df = pd.DataFrame(my_dict, 
                  columns=['학과','이름','학점','학년'], 
                  index=['one','two','three','four'])
display(df) 

#column indexing
print(df['이름'])    # series로 return
# one       홍길동
# two       아이유
# three     김연아
# four     신사임당
# Name: 이름, dtype: object

#print(df['이름':'학년'])   #error, column은 slicing이 안됨
display(df[['학과', '학년']])
```

```python
# row indexing (숫자 index를 사용)

# print(df[0])    # error, 단일 indexing 안됨
display(df[:2])     # 숫자인덱스로 slicing은 가능 -> dataframe형태로 나옴
# display(df[[0,2]])  # error, fancy indexing 안됨

# row indexing(지정 index를 사용)
# print(df['one'])    #error, column으로 인식해서 에러가 남
display(df['one':'three'])  #슬라이싱 가능. 결과는 dataframe으로나옴
#display(df[['one','three']])  #error, fancy indexing 안됨
```

## df.loc[ ]

    행과 열에 대한 indexing을 할 수 있음
    숫자 인덱스는 못쓰고 지정 인덱스만 사용가능!

```python
# df.loc[] 에 대해서 알아보자

my_dict = {'이름':['홍길동','아이유','김연아','신사임당'],
           '학과': ['철학','수학','컴퓨터','국어국문'],
           '학년':[1,3,2,4],
           '학점':[1.4,2.7,3.5,2.9]}

df = pd.DataFrame(my_dict, 
                  columns=['학과','이름','학점','학년'], 
                  index=['one','two','three','four'])
display(df)

print(df.loc['one'])  # 가능! 결과는 series로 나옴, 단일 row 추출 가능
display(df.loc['one':'three'])  # 가능, dataframe으로 결과가 나옴
display(df.loc[['one','three']]) # fancy indexing도 가능

# print(df.loc[0])      # error, 숫자 인덱스는 사용 불가
print(df.iloc[0])       # iloc는 숫자인덱스 가능
```

```python
#loc를 이요해서 row indexing 과 column indexing을 같이 할 수 있음
# df.loc[행,열]

display(df.loc['two':'three','학과':'이름'])
print(df.loc['two':'three','이름'])
display(df.loc['two':'three',['학년','이름']])
```

```python
# boolean indexing
# 학점이 3.0 이상인 학생의 학과와 이름을 출력하세요
my_dict = {'이름':['홍길동','아이유','김연아','신사임당'],
           '학과': ['철학','수학','컴퓨터','국어국문'],
           '학년':[1,3,2,4],
           '학점':[1.4,2.7,3.5,2.9]}

df = pd.DataFrame(my_dict, 
                  columns=['학과','이름','학점','학년'], 
                  index=['one','two','three','four'])
display(df)

#boolean mask를 생성
best = (df['학점'] >= 3.0)

display(df[best].loc[:,['학년','이름']])
display(df.loc[best,['학년','이름']])
```

```python
df = pd.read_csv('C:/jupyter_home/data/auto-mpg.csv', header=None)

df.columns = ['mpg','cylinders','displacement','horsepower','weight','acceleration','year','origin','name']
display(df)
```

```python
# info() ; dataframe의 기본 정보 나옴
print(df.info())
```

```python
# count(): 유효한 값의 개수(NaN이 아닌 값)
print(df.count())
```

```python
# value_counts() : series에 대해서 unique value의 개수를 알려줌
print(df['origin'].value_counts())    # 결과가 series로 나옴
# 1    249
# 3     79
# 2     70
# Name: origin, dtype: int64

# 원래 nan 값이 있으면 value_count에 포함되서 계산
# dropna를 True로 라면 NaN을 제외하고 수행함 -> 이 데이터는 결측치가 없어서 위에랑 결과가 같음
print(df['origin'].value_counts(dropna=True))
# 1    249
# 3     79
# 2     70
# Name: origin, dtype: int64
```

```python
# unique() : Series에 대해 중복해서 유일한 값이 어떤 값이 있는지 알려줌
print(df['year'].unique())   #[70 71 72 73 74 75 76 77 78 79 80 81 82]
```

```python
# isin(): boolean mask를 만들기 위해 많이 사용
df['origin'].isin([3])   # 제조국이 3인 mask
# isin()을 이용한 boolean indexing
df.loc[df['origin'].isin([3]),:] 
```

## 정렬 


```python
# 난수의 재현성을 확보하기 위해
np.random.seed(1)
df = pd.DataFrame(np.random.randint(0,10,(6,4)),
                  columns=['A', 'B', 'C', 'D'],
                  index=pd.date_range('20220101', periods=6))
display(df)
```

```python
# index를 랜덤하게 섞기
# np.random.shuffle(df.index)   # shuffle()은 원본데이터를 변경
                              # error! df의 인덱스 자체를 변경X (mutable operation 지원X)
 # 인덱스를 섞으려면 permutation() 사용
random_index = np.random.permutation(df.index)
print(random_index) #['2022-01-05T00:00:00.000000000' '2022-01-04T00:00:00.000000000'
                    #  '2022-01-03T00:00:00.000000000' '2022-01-02T00:00:00.000000000'
                    #  '2022-01-06T00:00:00.000000000' '2022-01-01T00:00:00.000000000']

# 변경된 인덱스로 dataframe 재설정
df2 = df.reindex(index=random_index,
                 columns=['B','A','D','C'])
display(df2)
```

```python
# 섞었으니까 정렬해보자

#index를 기반으로 정렬
display(df2.sort_index(axis=1, ascending=True))  # 컬럼을 정렬
display(df2.sort_index(axis=0, ascending=True))  # 행을 정렬

# value를 기반으로 정렬
display(df2.sort_values(by='B', ascending=True))  # 특정 컬럼의 값으로 행을 정렬
```
