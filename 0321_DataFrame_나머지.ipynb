```python
# dataframe 연결

import numpy as np
import pandas as pd

df1 = pd.DataFrame({'a':['a0','a1','a2','a3'],
                    'b':[1,2,3,4],
                    'c':['c0','c1','c2','c3']},
                   index=[0,1,2,3])
display(df1)

df2 = pd.DataFrame({'b':[5,6,7,8],
                    'c':['c0','c1','c2','c3'],
                    'd':['d0','d1','d2','d3'],
                    'e':['e0','e1','e2','e3']},
                   index=[2,3,4,5])
display(df2)

result_df = pd.concat([df1,df2],axis=0)
display(result_df)

result_df1 = pd.concat([df1,df2],axis=1)
display(result_df1)
```

# dataframe 결합(merge)

```python
#inner join
data1 = {'학번':[1,2,3,4],
         '이름':['홍길동','김연아','아이유','신사임당'],
         '학년':[3,2,1,4]}
df1= pd.DataFrame(data1)
display(df1)

data2 = {'학번':[1,2,3,4],
         '학과':['철학과','수학과','컴퓨터','국어국문'],
         '학점':[1.2,3.3,2.7,4.0]}
df2 = pd.DataFrame(data2)
display(df2)

result = pd.merge(df1,df2, on='학번', how='inner')   # inner join
display(result)
```

```python
data1 = {'학번':[1,2,3,4],
         '이름':['홍길동','김연아','아이유','신사임당'],
         '학년':[3,2,1,4]}
df1= pd.DataFrame(data1)
display(df1)

data2 = {'학번':[1,2,3,5],   # 5번이 들어옴!
         '학과':['철학과','수학과','컴퓨터','국어국문'],
         '학점':[1.2,3.3,2.7,4.0]}
df2 = pd.DataFrame(data2)
display(df2)

result = pd.merge(df1,df2, on='학번', how='inner')   # inner join
display(result)   # inner join 이기 때문에 겹치는 것만 출력
```

```python
# outer join
data1 = {'학번':[1,2,3,4],
         '이름':['홍길동','김연아','아이유','신사임당'],
         '학년':[3,2,1,4]}
df1= pd.DataFrame(data1)

data2 = {'학번':[1,2,3,5],   # 5번이 들어옴!
         '학과':['철학과','수학과','컴퓨터','국어국문'],
         '학점':[1.2,3.3,2.7,4.0]}
df2 = pd.DataFrame(data2)

result = pd.merge(df1,df2, on='학번', how='outer')   # outer join
display(result)   # outer join 이기 때문에 겹치치 않아도 포함

result = pd.merge(df1,df2, on='학번', how='left')   # left outer join
display(result)   # 왼쪽(df1)에 있는 것만 포함

result = pd.merge(df1,df2, on='학번', how='right')   # right outer join
display(result)   # 오른쪽(df2)에 있는 것만 포함
```

```python
# 결합 기준 컬럼명이 다른 경우  
data1 = {'학번':[1,2,3,4],
         '이름':['홍길동','김연아','아이유','신사임당'],
         '학년':[3,2,1,4]}
df1= pd.DataFrame(data1)

data2 = {'학생학번':[1,2,3,5],   # 5번이 들어옴!
         '학과':['철학과','수학과','컴퓨터','국어국문'],
         '학점':[1.2,3.3,2.7,4.0]}
df2 = pd.DataFrame(data2)

result = pd.merge(df1,df2, left_on='학번', right_on='학생학번', how='outer')   
display(result)  
```

```python
# 기준 컬럼과 다른 dataframe의 행 인덱스를 결합하려면?

data1 = {'학번':[1,2,3,4],
         '이름':['홍길동','김연아','아이유','신사임당'],
         '학년':[3,2,1,4]}
df1= pd.DataFrame(data1)

data2 = {'학과':['철학과','수학과','컴퓨터','국어국문'],
         '학점':[1.2,3.3,2.7,4.0]}
df2 = pd.DataFrame(data2,index=[1,2,3,5])

result = pd.merge(df1,df2, left_on='학번', right_index=True, how='outer')   
display(result)  
```

```python
# dataframe 행과 열에 함수를 mapping

titanic = sns.load_dataset('titanic')
display(titanic.head())

df = titanic.loc[:,['age','fare']]
display(df.head())

#함수는 apply()사용 (axis를 명시해야함)

# 정규화의 min_max scailing 처리
def min_max(s):
    return (s-s.min())/(s.max()-s.min())

# dataframe에 apply를 적용할때 axis=0하면 열단위
result = df.apply(min_max,axis=0)
display(result.head()) #dataframe에 대한 정규화 진행

def add_two_num(a,b):
    return a+b

result['add'] = df.apply(lambda x: add_two_num(x['age'],x['fare']), axis=1) #행단위
display(result.head())
```

# grouping

    복잡한 데이터를 어떤 기준에 따라서 여러 그룹으로 나눠서 관찰하기 위한 방법

```python
# dataframe의 1개의 열을 기준으로 grouping
titanic = sns.load_dataset('titanic')
df = titanic.loc[:,['age','sex','class','fare','survived']]
display(df.head())

grouped = df.groupby(['class'])

# group으로 묶인 모든 group 확인
for key, group in grouped:
    print('key', key)
    display(group.head(3))
```

```python
# 특정 그룹을 가져오기
grouped.get_group('Third')
```

```python
# 일단 group으로 묶이게 되면, 각 group에 대한 집계함수 사용 가능
my_mean = grouped.mean()
display(my_mean)    # 숫자로 된것만 평균을 냄
```
