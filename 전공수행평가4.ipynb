```python
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split 
import tensorflow as tf
from tensorflow.keras.layers import Flatten, Dense
from tensorflow.keras.optimizers import SGD
from sklearn.preprocessing import MinMaxScaler
from tensorflow.keras.models import Sequential
import warnings

# 경고메시지 출력하지 않아요!
warnings.filterwarnings(action='ignore')
```

```python
# data loading 
train = pd.read_csv('./data/train.csv')
display(train)
raw_test = pd.read_csv('./data/test.csv')
display(raw_test)
```

```python
train = train.set_index('PassengerId')
display(train)
test = raw_test.set_index('PassengerId')
display(train)
```

```python
# 필요없는 column(feature)는 삭제처리
train_data = train.drop(['Name','Ticket','Fare','Cabin'],
            axis=1, inplace=False)

test_data = test.drop(['Name','Ticket','Fare','Cabin'],
            axis=1, inplace=False)

display(train_data.head())
display(test_data.head())
```

```python
# SibSp = 함께 탄 형제,배우자 Parch= 함께 탄 부모,자식
# 두 칼럼은 하나로 합칠거임 => 동승자 수
train_data['Family'] = train_data['SibSp'] + train_data['Parch']
train_data.drop(['SibSp','Parch'],axis=1, inplace=True)

test_data['Family'] = test_data['SibSp']+test_data['Parch']
test_data.drop(['SibSp','Parch'],axis=1, inplace=True)

display(train_data)
display(test_data)
```

```python
# 결측치 확인
print(train_data.isnull().sum())
print(test_data.isnull().sum())
```

```python
# embarked 에 결측치가 2개임 최빈값을 이용해서 채움
train_data['Embarked'].fillna(train_data['Embarked'].mode()[0],inplace=True)
```

```python
# Age 결측치는 평균으로 채울거임
train_data['Age'] = train_data['Age'].fillna(train_data['Age'].mean())

test_data['Age'] = test_data['Age'].fillna(test_data['Age'].mean())
```

```python
# 결측치 확인
print(train_data.isnull().sum())
print(test_data.isnull().sum())
```

```python
# 문자로 되어있는 값은 숫자로 변경
gender_string={'male':0,'female':1}
train_data['Sex'] = train_data['Sex'].map(gender_string)

embarked_string = {'S':0,'C':1,'Q':2}
train_data['Embarked'] = train_data['Embarked'].map(embarked_string)

test_data['Sex'] = test_data['Sex'].map(gender_string)
test_data['Embarked'] = test_data['Embarked'].map(embarked_string)

display(train_data.head())
display(test_data.head())
```

```python
# age 칼럼을 카데고리화 하기
def age_category(age):
    if((age>=0) & (age<25)):
        return 0
    elif ((age>=25) & (age<50)):
        return 1
    else:
        return 2
    
train_data['Age'] = train_data['Age'].map(age_category)
test_data['Age'] = test_data['Age'].map(age_category)

display(train_data.head())
display(test_data.head())
```

```python
# 결측치 확인
print(train_data.isnull().sum())
print(test_data.isnull().sum())
```

```python
train_x_data = train_data.drop('Survived',axis=1,inplace=False)
train_t_data = train_data['Survived'].values.reshape(-1,1)
# 정규화
scaler = MinMaxScaler()
scaler.fit(train_x_data)
scaler.fit(test_data)
norm_train_x_data = scaler.transform(train_x_data)
norm_test_x_data = scaler.transform(test_data)
```

```python
# 이제 tensorflow로 돌려보기!
## placeholder
X = tf.placeholder(shape=[None,5],dtype=tf.float32)
T = tf.placeholder(shape=[None,1],dtype=tf.float32)

# weight & bias
W = tf.Variable(tf.random.normal([5,1]))
b = tf.Variable(tf.random.normal([1]))

# hypothesis, model, predict model, logistic regression model
logit = tf.matmul(X,W) + b
H = tf.sigmoid(logit)

# cross entropy(loss function)
loss = tf.reduce_mean(tf.nn.sigmoid_cross_entropy_with_logits(logits=logit, labels=T))

# train
train = tf.train.GradientDescentOptimizer(learning_rate=1e-4).minimize(loss)

#session
sess = tf.Session()
sess.run(tf.global_variables_initializer())

# 반복학습
# 전체 데이터를 이용해서 1한 학습 => 1 epoch(에폭)
for step in range(300000):
    _, loss_val = sess.run([train, loss], feed_dict={X:norm_train_x_data, T:train_t_data})
    if step % 30000 == 0:
        print('loss value : {}'.format(loss_val))
```

```python
predict = tf.cast(H >= 0.5, dtype=tf.int32)  
result = sess.run(predict, feed_dict={X:norm_test_x_data})
result_df = pd.DataFrame(result)
result_df['PassengerId'] = raw_test['PassengerId']
result_df.rename(columns = {0:'Survived'},inplace=True)
result_df = result_df[['PassengerId','Survived']]
result_df = result_df.set_index('PassengerId')
display(result_df)
```

```python
result_df.to_csv('./data/submission.csv')
```
