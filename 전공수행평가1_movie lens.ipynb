```python
import pandas as pd
import numpy as np
```

```python
# Movie Lens Data set을 이용(영화에 대한 평점을 준 정보)

movies = pd.read_csv('C:/jupyter_home/data/movies.csv')
ratings = pd.read_csv('C:/jupyter_home/data/ratings.csv')
display(movies)
display(ratings)
```

```python
df = pd.merge(movies,ratings, on='movieId', how='inner')   # inner join
display(df.head())
```

```python
df.info()
```

```python
# 1. 사용자가 평가한 모든 영화의 전체 평균 평점을 출력
result = ratings['rating'].mean()
print(result)
```

```python
# 2. 각 사용자 별 평균평점을 구하라 (userId로 오름차순)
grouped = ratings.groupby('userId')
result = grouped.mean()
print(result['rating'])
```

```python
# 각 영화별 평균 평점을 구하세요 (movieId로 오름차순)
grouped = df.groupby(['movieId','title'])
result = grouped.mean()
display(result['rating'])
```

```python
# 4. 평균 평점이 가장 높은 영화의 제목을 출력하세요.
#    단, 동률이 있을 경우 모두 출력하고 
#    title을 기준으로 오름차순 정렬하세요.

grouped = df.groupby(['movieId','title'])
result = grouped.mean()
result.sort_values(by=['rating','title'], ascending=[False,True], inplace=True)
max_score = result.iloc[0]['rating']
mask = result['rating']==max_score
print(result[mask]['rating'])
```

```python
# 5. Comedy영화 중 가장 평점이 낮은 영화의 제목을 출력하세요.
#    단, 동률이 있을 경우 모두 출력하고 
#    title을 기준으로 오름차순 정렬하세요.
df_comedy = df[df['genres'].str.contains('Comedy')]
grouped = df_comedy.groupby(['movieId','title','genres'])
result = grouped.mean()
result.sort_values(by=['rating','title'], ascending=[True,True], inplace=True)
min_score = result.iloc[0]['rating']
mask = result['rating']==min_score
print(result[mask]['rating'])
```

```python
import datetime
year=[]

for i in range(100836):
    year.append(datetime.datetime.fromtimestamp(df.iloc[i]['timestamp']).strftime('%Y'))

df['year'] = year
display(df.head())
```

```python
# 6. 2015년도에 평가된 모든 Romance 영화의 평균 평점은?
df_2015 = df[df['year'].str.contains('2015')]
df_romance = df_2015[df_2015['genres'].str.contains('Romance')]
print(df_romance['rating'].mean())
```

```python
# 7. 모든 영화장르 중 사용자 평점이 가장 좋은 
##    영화장르는 무엇인가요?
##    동률이 있으면 영화장르를 기준으로 오름차순 정렬하세요.
grouped = df.groupby(['genres'])
result = grouped.mean()
result.sort_values(by=['rating','genres'], ascending=[False,True], inplace=True)
max_score = result.iloc[0]['rating']
mask = result['rating']==max_score
print(result[mask]['rating'])
```
