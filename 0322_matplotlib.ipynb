## line chart(선그래프)

```python
import matplotlib.pyplot as plt

# plot 제목 설정
plt.title('Line Plot')

# plot()이라는 메소드로 line plot을 그릴 수있음
# x축의 눈금(tick)은 기본적으로 0,1,2,3,,,자동 생성, y축만 지정!
#plt.plot([1,5,12,25])

# tick을 별도로 지정하고 싶다면 아래처럼
plt.plot([10,20,30,40],[1,5,12,25])

# show로 화면에 rendering 하고 마우스 이벤트를 기다리는 함수
# 주피터노트북은 예외로 안써도 나옴
plt.show()
```

```python
import numpy as np
import pandas as pd

df = pd.read_excel('C:/jupyter_home/data/lineplot_sample_data.xlsx')
display(df.head())
```

```python
# NaN을 채우기 위해 이전 행의 값으로 채우기
df = df.fillna(method='ffill')
display(df.head())
```

```python
# 관심있는 데이터만 추리기
# 서울특별시에서 다른 지역으로 이동한 데이터만 추출

# boolean indexing 사용
mask = (df['전출지별']=='서울특별시') & (df['전입지별']!='서울특별시')
print(mask)

df_seoul = df.loc[mask,:]
display(df_seoul)
```

```python
# 전출지가 서울인것만 골랐으므로 전출지별 칼럼 삭제
df_seoul.drop('전출지별', axis=1,inplace=True)
display(df_seoul)
```

```python
# 전입지별 -> 전입지로 이름 변경
df_seoul.rename({'전입지별':'전입지'}, axis=1, inplace=True)
display(df_seoul)
```

```python
# 전입지를 index로 설정
df_seoul.set_index('전입지',inplace=True)
display(df_seoul)
```

```python
# 서울에서 경기도로 전입한 데이터만 가져오기
s = df_seoul.loc['경기도']
print(s)

plt.title('서울에서 경기도로 전입한 사람 추이')  # 한글깨짐
plt.plot(s.index, s.values)
plt.show()
```

```python
# 한글 폰트가 있어야 깨짐현상이 없애기
import matplotlib as mpl
import matplotlib.font_manager as fm
from matplotlib import rc

import warnings
warnings.filterwarnings(action='ignore') # 경고메시지 출력되지 않게 설정

# 그래프에서 '-' 기호때문에 문제가 발생할 여지가 있음
mpl.rcParams['axes.unicode_minus'] = False

# 폰트 사용하기
font_path = 'C:/jupyter_home/font/malgun.ttf'
font_name = fm.FontProperties(fname=font_path).get_name()
rc('font', family=font_name)

# 이전 셀에서 만들어둔 series를 이용해서 line plot 그리기
plt.title('서울에서 경기도로 전입한 사람 추이') 
plt.plot(s.index, s.values)
plt.xlabel('연도')
plt.ylabel('이동 인구수')
plt.show()
```

```python
# 이번에는 약간의 스타일 적용하기
plt.style.use('ggplot')

# 그래프 크기를 설정
plt.figure(figsize=(10,5))  # 가로세로 inch기준

# 연도를 보기 위해 label를 세로로 돌리기
plt.xticks(rotation='vertical')

plt.title('서울에서 경기도로 전입한 사람 추이') 
plt.plot(s.index, s.values,)
plt.xlabel('연도')
plt.ylabel('이동 인구수')

plt.show()
```

```python
plt.style.use('ggplot')
plt.figure(figsize=(10,5))  # 가로세로 inch기준
plt.xticks(rotation='vertical')

# 마커 찍기
plt.title('서울에서 경기도로 전입한 사람 추이') 
plt.plot(s.index, s.values, marker='o',markersize=5,markerfacecolor='r', color='g', linewidth=2)
plt.xlabel('연도')
plt.ylabel('이동 인구수')

# 범례
plt.legend(labels=['서울->경기'], loc='best')

plt.show()
```

```python
# sub plot => 그래프 여러개를 한번에 그리는 방법

# figure는 canvas에 나타내는 객체, 도화지 역할
# but, 주피터 노트북은 plot하면 자동으로 figure객체를 만들어줌

# figure 객체는 1개 이상의 axes로 구성
# plot 함수는 axes도 자동으로 생성
# 여러개의 axes를 만들어서 subplot을 만들거임
```

```python
plt.style.use('ggplot')
fig = plt.figure(figsize=(10,5))

# axes 만들기
ax1 = fig.add_subplot(1,2,1) # 1행2열의 첫번째 axes
ax2 = fig.add_subplot(1,2,2) # 1행2열의 두번째 axes

ax1.set_title('서울->경기 인구 이동')
ax1.set_xlabel('연도')
ax1.set_ylabel('이동 인구수')
ax1.plot(s.index,s.values)

ax2.plot(s.index, s.values, marker='o',markersize=5,markerfacecolor='r', color='g', linewidth=2)

plt.show()
```

```python
df = pd.read_csv('C:/jupyter_home/data/auto-mpg.csv', header=None)

df.columns = ['mpg','cylinders','displacement','horsepower','weight','acceleration','year','origin','name']
display(df.head(3))

# mpg 칼럼의 최대와 최소값
print(df['mpg'].max(),df['mpg'].min())

# matplot말고 pandas의 기능을 이용해서 히스토그램 그려보기
# mpg 칼럼의 최대와 최소사이를 10개의 구간으로 나눔
df['mpg'].plot(kind='hist', bins=10,color='blue',figsize=(10,5))
plt.show()
```

```python
# scatter
df = pd.read_csv('C:/jupyter_home/data/auto-mpg.csv', header=None)
df.columns = ['mpg','cylinders','displacement','horsepower','weight','acceleration','year','origin','name']
display(df.head(3))

# 두개의 변수를 사용해서 연비(mpg)와 weight(중량)을 이용해서 산점도를 그릴거임
df.plot(kind='scatter',
       x='weight',
       y='mpg',
       color='red',
       s=10,
       figsize=(10,5))
```

```python
# boxplot
# 단변수의 데이터분포를 살펴보기 위해 사용
# 이상치를 눈으로 확인할수 있음
df = pd.read_csv('C:/jupyter_home/data/auto-mpg.csv', header=None)
df.columns = ['mpg','cylinders','displacement','horsepower','weight','acceleration','year','origin','name']
display(df.head(3))

plt.boxplot(x=[df.loc[df['origin']==1,'mpg'],
              df.loc[df['origin']==2,'mpg'],
              df.loc[df['origin']==3,'mpg']])
plt.show()
```
